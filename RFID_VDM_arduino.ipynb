```python
#getting and working with data
import pandas as pd
import numpy as np
import re
import os
import datetime as dt
import string
import serial
import time

#visualizing results
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_context("poster")
sns.set_style("ticks")

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 15000)
pd.set_option('display.max_colwidth', -1)

import warnings; warnings.simplefilter('ignore')
np.set_printoptions(suppress=True)
```

```python
data = pd.DataFrame(columns=['timestamp', 'device', 'signal'])

ser = serial.Serial('/dev/cu.usbserial-DA014CBL', 9600)

start = dt.datetime.now()

while (1):
    #get time
    
    #read data
    a = ser.readline()
    timestamp = dt.datetime.fromtimestamp(time.time()).strftime('%Y-%m-%d %H:%M:%S')
    device = str(a)[2:4]
    signal = float(str(a).split(": ")[1].split('\\')[0])
    
    #only save data when photobeam in broken
    if signal > 70:
        data_int = pd.DataFrame(data=[timestamp, device, signal]).T
        data_int.columns = ['timestamp', 'device', 'signal']
        data = data.append(data_int, ignore_index=True)
    
    #check time and save every hour
    now = dt.datetime.now()
    diff = round((now - start).total_seconds())
    if diff % 60 == 0:
        print('saving data')
        data.to_csv('data.csv')
```
